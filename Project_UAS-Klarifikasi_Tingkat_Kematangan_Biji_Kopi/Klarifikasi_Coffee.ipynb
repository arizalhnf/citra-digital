```python
import matplotlib, cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```python
img = cv2.imread('./trainning/L.jpg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
height, width = img.shape[0:2]
startRow = int(height*.100)
startCol = int(width*.100)
endRow = int(height*.90)
endCol = int(width*.90)
crop = img[startRow:endRow, startCol:endCol]
cv2.imwrite('./testing/crop/light/L1.jpg', crop)
```

```python
L1 = cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)

plt.imshow(L1)
```

```python
gray_img = cv2.cvtColor(L1, cv2.COLOR_RGB2GRAY)

print(gray_img)
```

```python
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB))
```

```python
average_color_per_row = np.average(img, axis=0)
average_color = np.average(average_color_per_row, axis=0)
average_color = np.uint8(average_color)
print(average_color)
```

```python
average_color_img = np.array([[average_color]*100]*100, np.uint8)
plt.imshow(average_color_img)
```
